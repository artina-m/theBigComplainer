### Data Preprocessing

```python
import pandas
df = pandas.read_csv("311_Service_Requests.csv")
```

```python
df.head()
```

#### Filter down to top 10 complaints in Manhatten

```python
top10 = ["Noise - Residential", "HEAT/HOT WATER", "Illegal Parking", "Blocked Driveway",
        "Street Condition", "Street Light Condition", "UNSANITARY CONDITION",
        "Noise - Street/Sidewalk", "Water System"]
```

```python
man = df.loc[df['Borough'] == 'MANHATTAN']
top10complaints = man.loc[man['Complaint Type'].isin(top10)]
```

```python
len(top10complaints)
```

```python
top10complaintsDrop = top10complaints.drop(['Community Board', 'Agency Name','Descriptor',
                      'Location Type', 'Community Board','X Coordinate (State Plane)', 'Location',
                      'Y Coordinate (State Plane)', "City", "Closed Date", "Borough"], axis=1);
top10complaintsDrop =  top10complaintsDrop.dropna(axis=0, how='any') 
```

```python
len(top10complaintsDrop)
```

#### Add NYC Neighborhoods

```python
# https://www.health.ny.gov/statistics/cancer/registry/appendix/neighborhoods.htm

n1 = [10026, 10027, 10030, 10037, 10039] #central harlem
n2 = [10001, 10011, 10018, 10019, 10020, 10036] # chelsea and clinton
n3 = [10029, 10035] # east harlem
n4 = [10010, 10016, 10017, 10022] # gramercy park and murray hill
n5 = [10012, 10013, 10014] #greenwhich village and soho
n6 = [10004, 10005, 10006, 10007, 10038, 10280] # lower manhatten
n7 = [10002, 10003, 10009] #lower east side
n8 = [10021, 10028, 10044, 10065, 10075, 10128] # upper east side
n9 = [10023, 10024, 10025] # upper west side
n10 = [10031, 10032, 10033, 10034, 10040] #inwood and washington heights
```

```python
neigh = [];
for index,row in top10complaintsDrop.iterrows():
    zc = int(row['Incident Zip'])
    if zc in n1: n = 'Central Harlem'
    elif zc in n2: n = "Chelsea and Clinton"
    elif zc in n3: n = "East Harlem"
    elif zc in n4: n = "Gramercy Park and Murray Hill"
    elif zc in n5: n = "Greenwhich Village and SoHo"
    elif zc in n6: n = "Lower Manhatten"
    elif zc in n7: n = "Lower East Side"
    elif zc in n8: n = "Upper East Side"
    elif zc in n9: n = "Upper West Side"
    elif zc in n10: n = "Inwood and Washington Heights"
    else: n = "Other"
        
    neigh.append(n)
```

```python
top10complaintsDrop['Neighborhood'] = neigh
```

```python
top10complaintsDrop.head()
```

```python
nData = top10complaintsDrop[top10complaintsDrop['Neighborhood'] != "Other"]
len(nData)
```

```python
# Write csv
nData.to_csv("ManhattanComplaints", sep='\t')
```
